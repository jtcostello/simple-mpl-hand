plot a super simple hand

```python
hand = SuperSimpleHand()
hand.draw(idx_flexion=0, mrs_flexion=0, thumb_flexion=0)

hand = SuperSimpleHand()
hand.draw(idx_flexion=0, mrs_flexion=0.5, thumb_flexion=0.5)
```

create an animation of the hand moving

```python
plt.close('all')

hand = SuperSimpleHand()
fig = hand.fig
hand.draw(idx_flexion=0, mrs_flexion=0, thumb_flexion=0)
def animate(i):
    hand.ax.clear()
    hand.draw(idx_flexion=i/20, mrs_flexion=i/20, thumb_flexion=i/20)
    # hand.draw(idx_flexion=0, mrs_flexion=i/20, thumb_flexion=i/20)
    return hand.ax

anim = FuncAnimation(fig, animate, frames=range(0, 20), interval=200)

# save the animation as a gif
# anim.save('hand.gif', writer='imagemagick')
```

plot a hand with joints

```python
plt.close('all')
hand = SimpleHand()
hand.draw()
plt.show()

hand = SimpleHand()
hand.set_flex(th=0, idx=0.2, mid=0.4, ri=0.6, pi=0.8)
hand.draw()
plt.show()
```

create an animation

```python
# create animation
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# make the subplot fill the figure
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)


hand = SimpleHand(fig, ax)
hand.draw()
def animate(i):
    hand.ax.clear()
    hand.set_flex(idx=max(min((i-6)/30.0, 1), 0),
                    mid=max(min((i-4)/30.0, 1), 0),
                    ri=max(min((i-2)/30.0, 1), 0),
                    pi=max(min((i-0)/30.0, 1), 0),
                    th=max(min((i-8)/30.0, 1), 0),)
    hand.draw()
    return hand.ax

from matplotlib.animation import FuncAnimation
anim = FuncAnimation(fig, animate, frames=range(0, 40), interval=100)

# anim.save('hand_v2.gif', writer='imagemagick')
```
