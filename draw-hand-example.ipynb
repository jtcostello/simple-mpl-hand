```python
%matplotlib widget
%load_ext autoreload
%autoreload 2

from simplehand import Hand
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

plot a hand

```python
hand = Hand()
hand.draw(idx_flexion=0, mrs_flexion=0, thumb_flexion=0)

hand = Hand()
hand.draw(idx_flexion=0, mrs_flexion=0.5, thumb_flexion=0.5)

```

create an animation of the hand moving

```python
plt.close('all')

hand = Hand()
fig = hand.fig
hand.draw(idx_flexion=0, mrs_flexion=0, thumb_flexion=0)
def animate(i):
    hand.ax.clear()
    hand.draw(idx_flexion=i/20, mrs_flexion=i/20, thumb_flexion=i/20)
    # hand.draw(idx_flexion=0, mrs_flexion=i/20, thumb_flexion=i/20)
    return hand.ax

anim = FuncAnimation(fig, animate, frames=range(0, 20), interval=200)

# save the animation as a gif
anim.save('hand.gif', writer='imagemagick')

```
